```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
netflix_df = pd.read_csv("data/netflix_data.csv")
```

```python
# Removing TV Shows
netflix_subset = netflix_df[netflix_df['type'] != "TV Show"]
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
short_movies = netflix_movies[netflix_movies['duration'] < 60]
```

```python
def color_map(movie):
    if movie['genre'] == "Documentaries":
        return "blue"
    elif movie['genre'] == "Children":
        return "green"
    elif movie['genre'] == "Stand-Up":
        return "yellow"
    else:
        return "green"
        
colors = netflix_movies.apply(color_map, axis=1).tolist()
```

```python
ax = netflix_movies.plot.scatter(figsize=(12,8),
    x='release_year',y='duration', color=colors, 
    xlabel="Release year", ylabel="Duration (min)",
    title="Movie Duration by Year of Release")
```
